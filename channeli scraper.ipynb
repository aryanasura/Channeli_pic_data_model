```python
url1="https://results.channeli.in/2020/placement/branch/115_students.html"
```

```python
print(url1)
    
```

```python
r1=requests.get(url1)
htmlcontent1=r1.content
```

```python
print(htmlcontent1)
```

```python
data=[]
soup1=BeautifulSoup(htmlcontent1,'html.parser')
t=soup1.find_all('table')
#print(t)
for y in t:
    tp=soup1.find_all('tbody')
    print(tp)
    rows=soup1.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele])
(data.pop(0))
#print(data)
```

```python
heading_tags = ["h2"]
for tags in soup1.find_all(heading_tags):
    csv_name=tags.text.strip()
```

```python
fields = ['S.No.','Name', 'Enrollment No.', 'Company', 'Accepted'] 
with open(csv_name, 'w') as f:
      
    # using csv.writer method from CSV package
    write = csv.writer(f)
      
    write.writerow(fields)
    write.writerows(data)
```
