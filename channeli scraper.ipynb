```python
def get_csv(url1):
    r1=requests.get(url1)
    htmlcontent1=r1.content
    data=[]
    soup1=BeautifulSoup(htmlcontent1,'html.parser')
    t=soup1.find_all('table')
    #print(t)
    for y in t:
        tp=soup1.find_all('tbody')
        #print(tp)
        rows=soup1.find_all('tr')
        for row in rows:
            cols = row.find_all('td')
            cols = [ele.text.strip() for ele in cols]
            data.append([ele for ele in cols if ele])
    (data.pop(0))
    #print(data)
    heading_tags = ["h2"]
    for tags in soup1.find_all(heading_tags):
        csv_name=tags.text.strip()
    #print(csv_name)
    fields = ['S.No.','Name', 'Enrollment No.', 'Company', 'Accepted'] 
    with open(csv_name, 'w') as f:

        # using csv.writer method from CSV package
        write = csv.writer(f)

        write.writerow(fields)
        write.writerows(data)

    
```

```python
def get_csv_intern(url1):
    r1=requests.get(url1)
    htmlcontent1=r1.content
    data=[]
    soup1=BeautifulSoup(htmlcontent1,'html.parser')
    t=soup1.find_all('table')
    #print(t)
    for y in t:
        tp=soup1.find_all('tbody')
        #print(tp)
        rows=soup1.find_all('tr')
        for row in rows:
            cols = row.find_all('td')
            cols = [ele.text.strip() for ele in cols]
            data.append([ele for ele in cols if ele])
    (data.pop(0))
    #print(data)
    heading_tags = ["h2"]
    for tags in soup1.find_all(heading_tags):
        csv_name=tags.text.strip()
        csv_name=csv_name.replace('Placement','Intern')
        print(csv_name)
    fields = ['S.No.','Name', 'Enrollment No.', 'Company', 'Accepted'] 
    with open(csv_name, 'w') as f:

        # using csv.writer method from CSV package
        write = csv.writer(f)

        write.writerow(fields)
        write.writerows(data)
```

# Extracting link on basis of students of a branch of each year for internship

```python
def get_all_links_for_internship(link_year):
    for x in link_year:
        link_1=link_year[x][1]
    #     print(link1)
        r=requests.get(link_1)
        htmlcontent=r.content
        soup=BeautifulSoup(htmlcontent,'html.parser')
        allanchortags=soup.find_all('a')
        all_links=[]

        for link in allanchortags:
                link_text="https://results.channeli.in/" + link.get('href')
                if(link_text[-13:]=='students.html'):
                    all_links.append(link_text)

        for x in all_links:
            print(x)
            get_csv_intern(x)
```

```python
get_all_links_for_internship(link_year)
```

# Extracting link on basis of students of a branch of each year for placement

```python
def get_all_links_for_placement(link_year):
    for x in link_year:
        link_0=link_year[x][0]
    #     print(link1)
        r=requests.get(link_0)
        htmlcontent=r.content
        soup=BeautifulSoup(htmlcontent,'html.parser')
        allanchortags=soup.find_all('a')
        all_links=[]

        for link in allanchortags:
                link_text="https://results.channeli.in/" + link.get('href')
                if(link_text[-13:]=='students.html'):
                    all_links.append(link_text)

        del(all_links[0:11])
        for x in all_links:
            print(x)
            get_csv(x)
```
